```python
import pandas as pd
import matplotlib
data=pd.read_csv('vgsales.csv')
data
```

```python
#which year has most sale
sale=data.groupby(['Year']).sum()['Global_Sales']
sale
```

```python
sale.plot(kind='bar')
```

```python
#which platform has most sale
platform=data.groupby(['Platform']).sum()['Global_Sales']
platform
```

```python
platform.plot(kind='bar')
```
