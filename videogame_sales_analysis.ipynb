```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=20,15
data=pd.read_csv('vgsales.csv')
data
```

```python
#which year has most sale
sale=data.groupby(['Year']).sum()['Global_Sales']
sale
```

```python
sale.plot(kind='bar')
```

```python
#which platform has most sale
platform=data.groupby(['Platform']).sum()['Global_Sales']
platform
```

```python
platform.plot(kind='bar')
```

```python
#which typ of game is most saled.
category=data.groupby(['Genre']).sum()['Global_Sales']
category
```

```python
category.plot(kind='pie',autopct='%1.1f%%')
```
